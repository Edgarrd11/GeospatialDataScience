# Spatial graphs
osmnx can query data from OpenStreetMap.

In this example we extract the street-based graph of Yoyogi Park, near our data from Tokyo.

```python
import osmnx
```

```python
graph = osmnx.graph_from_place("Yoyogi Park, Shibuya, Tokyo, Japan")
```

The code snippet above sends the query to the OpenStreetMap server to fetch the data.

```python
type(graph)
```

We can have a quick inspection of the structure of the graph with the plot_graph method:

```python
osmnx.plot_graph(graph)
```

The resultant graph object is actually a MultiDiGraph from networkx, a graph libray written in python.

The graph here is stored as a collection of 103 nodes (streets instersections) and 278 edges (streets) that connect them.

```python
len(graph.nodes)
```

```python
len(graph.edges)
```

Each of these elements can be queried to obtain more information such as the location and ID of a node.

```python
graph.nodes[1520546819]
```

The characteristics of an edge

```python
graph.edges[(1520546819, 3010293622, 0)]
```

Or how the different components of the graph relate to each other.

For example, what other nodes are directly connected to node 1520546819

```python
list(graph.adj[1520546819].keys())
```

Network are easy to represent in Python, and are one of the three main data structures in geographic data science.
